```python
# Function to clean text by removing digits and unnecessary symbols
def clean_text(text):
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text

# Apply the clean_text function to the "content" column
data['text'] = data['text'].apply(clean_text)

# Now, the "content" column is cleaned
data['text'][7]
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# Function to remove stopwords from text
def remove_stopwords(text):
    words = text.split()  # Split the text into words
    # Remove stopwords
    filtered_words = [word for word in words if word.lower() not in stopwords.words('english')]
    return ' '.join(filtered_words)  # Rejoin the words into a cleaned sentence

# Apply the remove_stopwords function to the "content" column
data['text'] = data['text'].apply(remove_stopwords)
data['text'][7]
```

```python
# Get length column for each text

data['text_length'] = data['text'].apply(len)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to the categorical column
data['target'] = label_encoder.fit_transform(data['category'])
```

```python
tokenizer = Tokenizer(num_words=50000,oov_token="out of vacabulary")
tokenizer.fit_on_texts(data['text'].values)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

## here average length is 1544 so let us take max length as 2000 it will be fine

```python
X = tokenizer.texts_to_sequences(data['text'].values)
X = pad_sequences(X, maxlen=2000)
```

```python
# Train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.1, random_state = 42)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state = 42)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,GlobalMaxPooling1D,Dropout, LSTM, Dense,Bidirectional,SpatialDropout1D
from tensorflow import keras
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras import regularizers
```

```python
model=Sequential()
model.add(Embedding(50000,100,input_length=2000))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(5, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs=10, batch_size=64,validation_split=0.1)
```

```python
# Evaluate on the testing set
loss, accuracy = model.evaluate(X_test, Y_test)
print(f'Model Accuracy: {accuracy * 100:.2f}%')
```
